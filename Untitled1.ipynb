```julia
function generate( r, N )
    xval = rand();
    x = xval;
    for _ in 2:N
        xval = r * xval * (1 .- xval)
        x = vcat(x, xval)
    end
    return x;
end

generate(r) = generate( r, 200 )
```

```julia
x = generate(3.75)
```

```julia
function plotLogistic( r, N )
    x=generate(r,N)
    plt = plot(x
        ,markershape=:square ,markercolor=:maroon
        ,linecolor=:pink ,lw=1
        ,label="r = " *string(r)
        ,xlabel="number of iterations"
        ,ylabel="x-values"
        ,ylim=(0,1)
    )
    return plt
end

plotLogistic(r) = plotLogistics( r, N )
```
