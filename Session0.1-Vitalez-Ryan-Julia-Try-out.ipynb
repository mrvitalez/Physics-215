```julia
# This code came from our discussion last Tuesday, Sept. 13
function generate(r, N)
    xval = rand();
    x = xval;
    for _ in 2:N
        xval = r * xval * (1 .- xval)
        x = vcat(x, xval)
    end
    return x;
end

generate(r) = generate(r, 200)
```

```julia
x = generate(3.75)
```
